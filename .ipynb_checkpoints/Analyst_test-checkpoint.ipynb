```python
_, ax = plt.subplots()
ax.scatter(strike, bid, s = 10, color = 'r', alpha = 0.75, label='bid IV')
ax.scatter(strike, ask, s = 10, color = 'b', alpha = 0.75, label='ask IV')
plt.legend()
```

Найдем средние значения точек ask и bid и построим график

```python
y = [abs(k+j)*0.5 for k, j in zip(bid, ask)]
_, ax = plt.subplots()
ax.scatter(strike, bid, s = 10, color = 'r', alpha = 0.75, label='bid IV')
ax.scatter(strike, ask, s = 10, color = 'b', alpha = 0.75, label='ask IV')
ax.plot(strike, y, linewidth=2, label='mean(ask, bid)')
plt.legend()
```

Построим полиномы 2-го и 3-го порядка для найденных средних значений. Для этого воспользуемся функцией polyfit библиотеки numpy.

```python
fp_2, residuals, rank, sv, rcond = np.polyfit(strike, y, 2, full=True)
fp_3, residuals, rank, sv, rcond = np.polyfit(strike, y, 3, full=True)
f_2 = np.poly1d(fp_2)
f_3 = np.poly1d(fp_3)

_, ax = plt.subplots()
ax.scatter(strike, bid, s = 10, color = 'r', alpha = 0.75, label='bid IV')
ax.scatter(strike, ask, s = 10, color = 'b', alpha = 0.75, label='ask IV')
ax.plot(strike, y, linewidth=1, color = 'slategrey', label='mean(ask, bid)')
ax.plot(strike, f_2(strike), linewidth=1, color='orange', label='2nd order polynom')
ax.plot(strike, f_3(strike), linewidth=1, color='crimson', label='3rd order polynom')
plt.legend()
```

Проверим, достаточно ли нам полинома 2-го порядка для приемлемого описания нашей выборки. Для этого найдем коэффициент детерминации для значений полиномов.

```python
mean_y = np.mean(ask+bid)
numerator_2 = numerator_3 = denominator = 0
for i in range(len(y)):
    numerator_2 += (ask[i] - f_2(strike[i])) ** 2
    numerator_2 += (bid[i] - f_2(strike[i])) ** 2
    numerator_3 += (ask[i] - f_3(strike[i])) ** 2
    numerator_3 += (bid[i] - f_3(strike[i])) ** 2
    denominator += (ask[i] - mean_y) ** 2
    denominator += (bid[i] - mean_y) ** 2
r_2 = 1 - numerator_2 / denominator
r_3 = 1 - numerator_3 / denominator
print("Коэффициент детерминации для значений полинома 2-го порядка:", r_2)
print("Коэффициент детерминации для значений полинома 3-го порядка:", r_3)
```

Мы не можем считать полином 2-го порядка приемлемым для описания нашей выборки, так как коэффициент детерминации для его значений < 0.5.

Мы должны ставить так, чтобы максимизировать вероятность.
Если казино нечестное, то имеет смысл ставить против всех.
Если честное, то мне известен только "Мартингейл"(сомнительный).
Что ж, будем удваивать ставку каждый раз. 
Цикл удвоения ставки до победы будем называть махинацией.
За одну махинацию мы выигрываем наш начальный процент банка.
Пусть по правилам казино ставить мы можем минимум 0.1% от банка.
Нужно понять, с какого процента нам начинать для того, 
чтобы вероятность не уехать в лес была максимальной.

```python
def probability(x,k):
# x - это ставка в процентах от банка
# k - во сколько раз мы хотим увеличить банк
# n - количество ставок, которые мы можем сделать в одной махинации, пока не проиграем
    n = int(np.log2(100/x + 1))
# prob - вероятность проиграть в одной махинации
    prob = (19/37) ** n
# num - количество махинаций, которое нам необходимо провести
    num = int(np.log(k) / np.log(1+x/100)) + 1
# ans - вероятность не уехать в лес
    return (1 - prob)**num
```

```python
max = 0
# x - это процент нашего банка
for x in np.arange(0.1, 100, 0.1):
    ans = probability(x, 10)
    if ans >= max:
        bet = x
        max = ans
print('Начальная ставка {bet}%, вероятность уехать в лес {max}%'.format(bet=float('{:.1f}'.format(bet)), max=float('{:.1f}'.format(100 - max*100))))
```

```python
martingale = probability(14.2, 2)
full_bank = (18/37)**1
print('Если пользуемся мартингейлом {martingale}%, если ставим весь банк {full_bank}%'.format(martingale=float('{:.1f}'.format(100 - martingale*100)), full_bank=float('{:.1f}'.format(100 - full_bank*100))))
```

Вычислим медиану стандартного отклонения приращений на скользящем окне 30 дней.

```python
def standard_deviation(data, start, end):
    mean_data = np.mean(data[start:end])
    sd = 0
    for i in range(start, end):
        sd += (data[i] - mean_data)**2
    return np.sqrt(sd/(end - start))
```

```python
stand_dev_day = list()
stand_dev_week = list()
for i in range(period-30):
    stand_dev_day.append(standard_deviation(change_per_day, i, i+30))
    stand_dev_week.append(standard_deviation(change_per_week, i, i+30))
print('Медиана стандартного отклонения дневных приращений: ', median(stand_dev_day))
print('Медиана стандартного отклонения недельных приращений: ', median(stand_dev_week))
```

Оценим показатель Хёрста для нашего ряда. 

```python
def Hurst(data):
    log_RS = list()
    log_n = list()
    period = 365
    for n in range(10, int(len(data)/2)):
        R = S = 0
        for i in range(0, period-n, n):
            max = -100
            min = 100
            mean_data = np.mean(data[i:i+n])
            sd = 0
            dev = 0
            for j in range(i, i+n):
                dev += data[j] - mean_data
                if dev > max:
                    max = dev
                if dev < min:
                    min = dev
                sd += (data[j] - mean_data)**2
            R += max - min
            S += np.sqrt(sd/n)
        log_RS.append(np.log(R/S))
        log_n.append(np.log(n))
    return np.polyfit(log_n, log_RS, 1)[0]
```

```python
H = Hurst(change_per_day)

print('Коэффициент Хёрста:', H)
```

Коэффициент Хёрста > 0.5, следовательно наш ряд сохраняет тенденцию, будущие данные опираются на прошлые. Этот показатель вместе с графиком распределения говорят нам о том, биткоин достаточно устойчив, чтобы рассмотреть его для покупки на коррекции.
