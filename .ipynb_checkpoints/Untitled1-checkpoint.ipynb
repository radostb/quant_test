```python
import pandas as pd
from statistics import mean
import numpy as np
```

```python
data = pd.read_excel('bitcoin-excel.xlsx', engine='openpyxl')
```

```python
data['Цена, USD']
```

```python
change_per_day = list()
change_per_week = list()

for i in range(len(data)-1):
    change_per_day.append((data['Цена, USD'][i] - data['Цена, USD'][i+1])*100/data['Цена, USD'][i+1])
for i in range(len(data)-7):    
    change_per_week.append((data['Цена, USD'][i] - data['Цена, USD'][i+7])*100/data['Цена, USD'][i+7])
for i in range(7):
    change_per_week.append(0)
change_per_day.append(0)
```

```python
data = data.assign(change_per_day = change_per_day)
data = data.assign(change_per_week = change_per_week)
```

```python
mean_per_day = mean(change_per_day)
mean_per_week = mean(change_per_week)
```

```python
m1_per_day = m2_per_day = m3_per_day = m1_per_week = m2_per_week = m3_per_week = 0
for i in range(len(data)):
    m1_per_day += (change_per_day[i] - mean_per_day)
    m2_per_day += (change_per_day[i] - mean_per_day)**2
    m3_per_day += (change_per_day[i] - mean_per_day)**3
    m1_per_week += (change_per_week[i] - mean_per_week)
    m2_per_week += (change_per_week[i] - mean_per_week)**2
    m3_per_week += (change_per_week[i] - mean_per_week)**3
m1_per_day /= len(data)
m2_per_day /= len(data)
m3_per_day /= len(data)
m1_per_week /= len(data)
m2_per_week /= len(data)
m3_per_week /= len(data)
```
